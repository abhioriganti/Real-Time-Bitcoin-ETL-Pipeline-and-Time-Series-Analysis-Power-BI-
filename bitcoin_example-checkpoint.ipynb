This tutorial demonstrates how to collect real-time Bitcoin data, visualize it using Power BI, and apply time series analysis in Python to forecast future price movements.

**In this notebook, we demonstrate how to::**
- Read and clean Bitcoin price data
- Visualize price, moving average, and volatility
- Forecast future prices using SARIMA model

####  Step 1: Load and Inspect the Data

In this step, we import the essential Python libraries for data handling, time series forecasting, and visualization. We use:
- `pandas` for data manipulation,
- `matplotlib` for plotting,
- `statsmodels` for SARIMA modeling,
- `warnings` to suppress irrelevant output warnings.

```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings("ignore")
```

#### Step 2: Load and Preprocess the Dataset

We load the cleaned Bitcoin price data from a CSV file. To maintain time continuity and avoid noise from outdated entries, we:
- Convert timestamps to datetime format,
- Drop the first 3 rows that don't follow the current time sequence,
- Sort by timestamp to ensure correct time series order,
- Set the timestamp as the index (SARIMA requires it),
- Calculate a 7-minute rolling average to smooth price fluctuations.

```python
# Loading the transformed dataset
df = pd.read_csv("bitcoin_price_transformed.csv")

# Converting timestamp to datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Removing old entries (drop first 3 rows)
df = df.iloc[3:].reset_index(drop=True)
df = df.sort_values(by='timestamp')
df.set_index('timestamp', inplace=True)

# Calculating 7-minute moving average
df['moving_avg_price'] = df['price_usd'].rolling(window=7).mean()

# Preview
df.head()
```

#### Step: Decompose Bitcoin Price Time Series
We use `seasonal_decompose` from statsmodels to break down the price data into trend, seasonality, and residuals. This helps us understand cyclical patterns and noise.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series
decomposition = seasonal_decompose(df['price_usd'], model='additive', period=60)
decomposition.plot()
plt.suptitle("Seasonal Decomposition of Bitcoin Price", fontsize=14)
plt.tight_layout()
plt.show()
```

#### Interpretation of Seasonal Decomposition

Using `seasonal_decompose`, we broke down the Bitcoin price time series into 3 components:

- **Trend**: Shows the overall upward/downward direction in price. We observed a mild declining trend in our data window.
- **Seasonality**: Represents periodic patterns within each hour. There was a repeating oscillation every few minutes, suggesting micro-trends likely caused by trading bots or short-term speculation.
- **Residuals**: Captures irregular or random noise in price movements.

This decomposition provides useful insight into recurring behavior and helps isolate unpredictable fluctuations from reliable trends.


####  Step 4: Building and Fitting the SARIMA Model

SARIMA (Seasonal AutoRegressive Integrated Moving Average) is used to model and forecast time series data.
We use `SARIMAX` from `statsmodels` to fit the model on our Bitcoin price data.
Here, the parameters `(1,1,1)x(1,1,1,12)` indicate:
- ARIMA part: p=1, d=1, q=1
- Seasonal part: P=1, D=1, Q=1, period=12 (12-minute cycle assumed)
You can tune these values for better results.

```python
# Fit SARIMA model (parameters can be tuned)
sarima_model = SARIMAX(df['price_usd'], order=(1,1,1), seasonal_order=(1,1,1,12))
sarima_fit = sarima_model.fit(disp=False)
```

#### Step 5: Forecasting Next 60 Minutes of Bitcoin Price

Once the model is trained, we forecast the next 60 time steps (minutes). 
We also generate future timestamps aligned with the forecast using `pandas.date_range` to visualize them chronologically.

```python
# Forecasting 60 future steps
forecast_steps = 60
forecast = sarima_fit.forecast(steps=forecast_steps)

# Creating future timestamps
last_timestamp = df.index[-1]
future_timestamps = pd.date_range(start=last_timestamp, periods=forecast_steps+1, freq='min')[1:]
```

#### Seasonal Decomposition Analysis

Using `seasonal_decompose()`, we broke down the Bitcoin price time series into:
- **Trend**: long-term price movement
- **Seasonality**: repetitive intra-hour patterns
- **Residual**: randomness or noise

This decomposition helps us visually detect periodic behavior and anomalies in the Bitcoin price data.

#### Step 6: Plotting Actual, Moving Average, and Forecasted Price

This visualization overlays three important components:
- Actual Bitcoin prices (historical),
- 7-minute moving average,
- SARIMA-based forecast for the next 60 minutes.
This provides a holistic view of both observed trends and predicted movement.

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['price_usd'], label='Actual Price')
plt.plot(df.index, df['moving_avg_price'], label='7-Minute Moving Average', linestyle='--', color='orange')
plt.plot(future_timestamps, forecast, label='Forecasted Price (SARIMA)', linestyle='--', color='green')
plt.title("Bitcoin Price Forecast (Next 60 Minutes) - SARIMA")
plt.xlabel("Time")
plt.ylabel("USD")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Step: Forecasting with Prophet
We use Facebooks Prophet model to forecast Bitcoin prices. Prophet is well-suited for time series with clear trends and seasonality.

```python
from prophet import Prophet

# Preparing data for Prophet
df_prophet = df.reset_index()[['timestamp', 'price_usd']]
df_prophet.columns = ['ds', 'y']

# Initializing and fitting the model
prophet_model = Prophet()
prophet_model.fit(df_prophet)

# Creating future dataframe
future = prophet_model.make_future_dataframe(periods=60, freq='min')
forecast_prophet = prophet_model.predict(future)

# Plotting forecast
fig1 = prophet_model.plot(forecast_prophet)
plt.title("Bitcoin Price Forecast (Prophet, Next 60 Minutes)")
plt.xlabel("Time")
plt.ylabel("Price (USD)")
plt.show()
```

#### Comparison: SARIMA vs Prophet

We used two forecasting models  **SARIMA** and **Prophet**  to predict Bitcoin price for the next 60 minutes.

- **SARIMA** is a classical statistical model that assumes a fixed seasonal cycle. It worked well on short-term historical trends and captured recurring patterns.
- **Prophet** by Facebook handles missing values and trend/seasonality shifts more robustly. It generated smooth future predictions even with slightly noisy data.

**Observation**: Prophet provided a smoother forecast while SARIMA captured short-term fluctuations more aggressively. The best model may depend on how volatile the Bitcoin market is in the selected period.
